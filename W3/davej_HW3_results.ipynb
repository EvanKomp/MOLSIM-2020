```python
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('ch4_opt.xlsx')
```

```python
df['Energy (kj/mol)'] = df['Energy (au)']*2625.5 # conversion from atomic units to kj/mol
```

```python
df
```

```python
labels = ['HG + STO-3G','HG + 3-21G*','HF + 6-31G**','B3LYP + 3-21G*','B3LYP + 6-31G**','B3LYP + DGDZVP']

fig, ax = plt.subplots(dpi=100)
ax.scatter(df['Num basis functions'].values,df['Time (s) '].values,c='darkturquoise')
for i, txt in enumerate(labels):
    ax.annotate(txt,(df['Num basis functions'][i],df['Time (s) '][i]),
                xytext=(df['Num basis functions'][i]+1.4,df['Time (s) '][i]))
    
ax.set_xlabel('Basis set size (# functions)',fontsize=15)
ax.set_ylabel('CPU time (seconds)',fontsize=15)
ax.set_title('CPU time versus basis set size', fontsize=18)
ax.set_facecolor('gainsboro')
```
